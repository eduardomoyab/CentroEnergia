```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge
```

```python
# Carga de datos
data = pd.read_excel('Transantiago.xlsx',sheet_name='Transantiago')
data = data[['Año','KM_Transantiago','vkm_rigido','vkm_articulado','Fraccion Rigidos']][2:]
```

```python
data4= data.copy()
data4 = data4[(data4['Año']!=2020)& (data4['Año']!=2021)]

data_pandemia = data[4:6]
data4
```

```python
scaler = StandardScaler()


# Separar los datos en entrenamiento y prueba
train_data = data4[data4['Año'] < 2025]
test_data = data4[data4['Año'] >= 2025]

# Seleccionar las variables independientes y dependiente
X_train = train_data[['vkm_rigido','vkm_articulado']]
y_train = train_data['Fraccion Rigidos']
X_test = test_data[['vkm_rigido','vkm_articulado']]
y_test = test_data['Fraccion Rigidos']

# Ajustar el objeto con los datos de entrenamiento
scaler.fit(X_train)

X_train_norm = scaler.transform(X_train)
X_test_norm = scaler.transform(X_test)



# Entrenar el modelo
model = Ridge()
model.fit(X_train_norm, y_train)

# Hacer predicciones
y_pred = model.predict(X_test_norm)

df_all = pd.concat([train_data, test_data])
df_all['Fraccion Rigidos'] = np.concatenate((y_train, y_pred))

df_all = pd.concat([df_all,data_pandemia])
df_all = df_all.sort_values('Año')
```

```python
df_all['KM_Transantiago'] = df_all['vkm_articulado']+df_all['vkm_rigido']*df_all['Fraccion Rigidos']
df_all
```

```python
df_aux = df_all[['Año','Fraccion Rigidos']]
df_aux = df_aux[df_aux['Año']>=2017]
df_aux.to_csv('Fracciones Rigidos.csv',index=False,encoding='utf-8')
```
